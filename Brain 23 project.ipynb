```python
# Defining dependent and independent varaibles
y = dataM["HC"]
X = dataM[['LSylvian3D1', 'RSylvian3D', 'LSuptemp3D1','RSuptemp3D']]
#Training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

# Logistic Regression

```python
logReg = LogisticRegression()
logReg.fit(X_test, y_test)
log_score = logReg.score(X_test, y_test)
print("The accuracy of the Logistic model is", log_score*100)
print('')
log_pred = logReg.predict(X_test)
print(classification_report(y_test, log_pred))
```

# Decision Tree classifier

```python
DCC = DecisionTreeClassifier()
DCC.fit(X_test, y_test)
dt_score = DCC.score(X_test, y_test)
print("The accuracy of Decision tree classifier model is", dt_score)
print('')
Dectree_pred = DCC.predict(X_test)
print(classification_report(y_test, Dectree_pred))
```

# Random Forest Classifier

```python
RFC = RandomForestClassifier(n_estimators = 2)
RFC.fit(X_test, y_test)
rf_score = RFC.score(X_test, y_test)
print("The accuracy of Random forest model is", rf_score)
print('')
randfor_pred = RFC.predict(X_test)
print(classification_report(y_test, randfor_pred))
```

# Gradient Boosting Classifier

```python
GBC = GradientBoostingClassifier(learning_rate=0.1)
GBC.fit(X_test, y_test)
Graboo_pred = GBC.predict(X_test)
gb_score = GBC.score(X_test, y_test)
print("The accuracy of Gradient boosting model is", gb_score)
print('')
print(classification_report(y_test, Graboo_pred))
```

# Gaussian Naive Bayes classifier

```python
GNB = GaussianNB()
GNB.fit(X_test, y_test)
nb_score = GNB.score(X_test, y_test)
print("The accuracy of Naive Bayes model is", nb_score)
print('')
Naibay_pred = GNB.predict(X_test)
print(classification_report(y_test, Naibay_pred))
```
